```python
x = np.linspace(-1,2,10)
f = 3*x*x+x-1


plt.figure(figsize=(3,3))

plt.plot(x,f, label="linija")
plt.scatter(x,f)
plt.ylabel('$f(x)$')
plt.xlabel("x")
plt.title("Grafik funkcije")
plt.legend()
plt.grid(True)
plt.show()
plt.savefig("plot1.png")
```

# Colourmesh

```python
C = np.array([[0,3,0],[1,2,1],[0,1,0]])
plt.pcolormesh(C)
plt.colorbar()
plt.show()
```

```python
plt.style.use('default')

def f(x, y):
    return np.sin(x*y)

x = np.linspace(- np.pi, np.pi, 100)
y = np.linspace(- np.pi, np.pi, 100)

xx, yy = np.meshgrid(x,y)
C2 = f(xx, yy)

plt.pcolormesh(xx, yy, C2)
plt.colorbar()
plt.show()
```

# Contour


```python
def g(x, y):
    return (x**2+y**2)**(-1/300)

C3 = g(xx, yy)

plt.contour(xx, yy, C3, 30)
plt.colorbar()
plt.show()
```

```python
plt.contourf(xx, yy, C3, 100)
plt.colorbar()
plt.show()
```

```python
x2 = np.linspace(-np.pi, np.pi, 30)
y2 = np.linspace(-np.pi, np.pi, 30)
xx2, yy2 = np.meshgrid(x2,y2)

U = np.sin(xx2)
V = np.cos(yy2)

C4 = np.linalg.norm(np.array([U,V]), axis = 0)
```

```python
plt.quiver(xx2, yy2, U, V, C4)
plt.colorbar()
plt.show()
```

# Subplot / Multiplot

```python
x = np.linspace(-3, 3, 100)

plt.subplot(221)
plt.plot(x,x)
plt.title("Subplot 1")

plt.subplot(222)
plt.plot(x,x*x)
plt.title("Subplot 2")

plt.subplot(223)
plt.plot(x,x**3)
plt.title("Subplot 3")

plt.subplot(224)
plt.plot(x,x*2+1)
plt.title("Subplot 4")

plt.show()
```
